```python

def visualize_average_wealth(billionaires_df):
    # Group by 'selfMade' and calculate average wealth
    average_wealth_by_type = billionaires_df.groupby('selfMade')['finalWorth'].mean()

    # Print the results
    print("Average wealth of self-made billionaires: {average_wealth_by_type['selfMade']:.2f}")
    print("Average wealth of inherited billionaires: {average_wealth_by_type['inherited']:.2f}")

    # Visualization: Bar plot for Average Wealth of Self-Made vs. Inherited Billionaires
    plt.figure(figsize=(8, 6))
    sns.barplot(x=average_wealth_by_type.index, y=average_wealth_by_type, palette='Set2')
    plt.title('Average Wealth of Self-Made vs. Inherited Billionaires')
    plt.xlabel('Billionaire Type')
    plt.ylabel('Average Wealth (in USD)')
    plt.show()

def visualize_top_billionaires_locations(billionaires_df):
    # Count the number of billionaires per city and nation
    city_counts = billionaires_df['city'].value_counts()
    nation_counts = billionaires_df['country'].value_counts()

    # Identify the city and nation with the most billionaires
    most_billionaires_city = city_counts.idxmax()
    most_billionaires_nation = nation_counts.idxmax()

    # Print the results
    str1 = (f"The city that has created the most billionaires is {most_billionaires_city} with {city_counts.max()} billionaires.")
    str2 = (f"The nation that has created the most billionaires is {most_billionaires_nation} with {nation_counts.max()} billionaires.")

    # Visualization: Bar plot for Top 10 Cities with the Most Billionaires
    plt.figure(figsize=(12, 8))
    top_cities = city_counts.head(10)
    sns.barplot(x=top_cities, y=top_cities.index, palette='viridis')
    plt.title('Top 10 Cities with the Most Billionaires')
    plt.xlabel('Number of Billionaires')
    plt.ylabel('City')
    plt.show()

    # Visualization: Bar plot for Top 10 Nations with the Most Billionaires
    plt.figure(figsize=(12, 8))
    top_nations = nation_counts.head(10)
    sns.barplot(x=top_nations, y=top_nations.index, palette='viridis')
    plt.title('Top 10 Nations with the Most Billionaires')
    plt.xlabel('Number of Billionaires')
    plt.ylabel('Nation')
    plt.show()
    return str1+str2
    
def visualize_self_made_billionaires(billionaires_df):
    # Step 1: Count the number of self-made billionaires
    self_made_counts = billionaires_df['selfMade'].value_counts()

    # Step 2: Print the results
    print(f"The number of self-made billionaires is {self_made_counts[0]}")

    # Step 3: Visualization - Bar plot for Self-Made vs. Inherited Billionaires
    fig, ax = plt.subplots(figsize=(8, 6))
    bars = ax.bar(self_made_counts.index, self_made_counts, color=['skyblue', 'lightcoral'])

    # Adding data labels on top of the bars
    for bar in bars:
        yval = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2, yval + 20, round(yval, 1), ha='center', va='bottom', color='black')

    # Adding title and labels
    plt.title('Self-Made vs. Inherited Billionaires')
    plt.xlabel('Billionaire Type')
    plt.ylabel('Number of Billionaires')

    # Display the plot
    plt.show()

# Call the function with your DataFrame
visualize_self_made_billionaires(billionaires_df)

def main():
    import pandas as pd
    import numpy as np
    from sklearn.impute import SimpleImputer
    import matplotlib.pyplot as plt
    import seaborn as sns
    import geopandas as gpd
    from mpl_toolkits.axes_grid1 import make_axes_locatable
    import plotly.express as px
    
    file_path = 'Billionaires_statistics_Dataset.csv'
    
    # Load the dataset into a DataFrame
    billionaires_df = pd.read_csv(file_path)
    
    # Display the first few rows of the dataset to verify the loading
    print(billionaires_df)
    
    billionaires_df.info()
    
    billionaires_df.isnull().sum()
    
    billionaires_df['cpi_country']
    
    # List of numerical columns
    numerical_cols = ['age', 'birthYear', 'cpi_country', 'gdp_country', 'gross_tertiary_education_enrollment',
          'gross_primary_education_enrollment_country', 'life_expectancy_country', 'tax_revenue_country_country',
          'total_tax_rate_country', 'population_country', 'latitude_country', 'longitude_country']
    
    # List of categorical columns
    categorical_cols = ['country', 'city', 'state']
    
    # List of date columns
    date_cols = ['birthDate']
    
    # Convert 'cpi_country' column to numeric format
    billionaires_df['gdp_country'] = billionaires_df['gdp_country'].replace('[\\$,]', '', regex=True).astype(float)
    
    # Impute missing values for numerical columns
    numerical_imputer = SimpleImputer(strategy='mean')
    billionaires_df[numerical_cols] = numerical_imputer.fit_transform(billionaires_df[numerical_cols])
    
    # Impute missing values for numerical columns
    #numerical_imputer = SimpleImputer(strategy='mean')
    #billionaires_df[numerical_cols] = numerical_imputer.fit_transform(billionaires_df[numerical_cols])
    
    # Impute missing values for categorical columns
    categorical_imputer = SimpleImputer(strategy='most_frequent')
    billionaires_df[categorical_cols] = categorical_imputer.fit_transform(billionaires_df[categorical_cols])
    
    # Convert 'birthDate' column to datetime format
    billionaires_df['birthDate'] = pd.to_datetime(billionaires_df['birthDate'], errors='coerce')
    
    # Impute missing values for date columns
    most_frequent_date = billionaires_df['birthDate'].mode()[0]
    billionaires_df['birthDate'] = billionaires_df['birthDate'].fillna(most_frequent_date)
    # Display columns with missing values after imputation
    
    # Impute missing values for 'birthMonth' using the mode (most frequent)
    mode_imputer_birth_month = SimpleImputer(strategy='most_frequent')
    billionaires_df['birthMonth'] = mode_imputer_birth_month.fit_transform(billionaires_df[['birthMonth']])
    
    # Impute missing values for remaining columns
    remaining_cols = ['residenceStateRegion', 'birthDay', 'cpi_change_country', 'organization', 'title']
    
    for col in remaining_cols:
        mode_imputer_remaining = SimpleImputer(strategy='most_frequent')
        # Use double square brackets to preserve DataFrame structure
        billionaires_df[[col]] = mode_imputer_remaining.fit_transform(billionaires_df[[col]])
    
    print(billionaires_df.isnull().sum())
    
    # Count the number of duplicated rows
    num_duplicates = billionaires_df.duplicated().sum()
    
    # Print the number of duplicated rows
    print(f"Number of duplicated rows: {num_duplicates}")
    
    #Distribution of Billionaires' Age:
    #Visualize the distribution of the age of billionaires using a histogram or a kernel density plot.
    plt.figure(figsize=(10, 6))
    sns.histplot(billionaires_df['age'], bins=20, kde=True)
    plt.title('Distribution of Billionaires\' Age')
    plt.xlabel('Age')
    plt.ylabel('Frequency')
    plt.show()
    
    #Geographic Distribution of Billionaires:
    #Create a scatter plot or a world map showing the geographic distribution of billionaires.
    world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
    billionaires_geo = gpd.GeoDataFrame(billionaires_df, geometry=gpd.points_from_xy(billionaires_df.longitude_country, billionaires_df.latitude_country))
    fig, ax = plt.subplots(figsize=(15, 10))
    world.plot(ax=ax, color='lightgray')
    billionaires_geo.plot(ax=ax, markersize=30, color='red', alpha=0.7)
    plt.title('Geographic Distribution of Billionaires')
    plt.show()
    
    #Wealth Distribution by Country:
    #Use a bar chart to visualize the total wealth of billionaires in each country.
    plt.figure(figsize=(18, 20))
    sns.barplot(x='finalWorth', y='country', data=billionaires_df.groupby('country')['finalWorth'].sum().reset_index().sort_values(by='finalWorth', ascending=False))
    plt.title('Wealth Distribution by Country')
    plt.xlabel('Total Wealth (USD)')
    plt.ylabel('Country')
    
    #Self-Made vs. Inherited Wealth:
    #Create a pie chart or a countplot to show the proportion or count of self-made and inherited billionaires.
    plt.figure(figsize=(8, 8))
    billionaires_df['selfMade'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['skyblue', 'lightcoral'])
    plt.title('Self-Made vs. Inherited Wealth')
    plt.ylabel('')
    plt.show()
    
    #Correlation Heatmap:
    #Visualize the correlation between numerical variables using a heatmap.
    
    
    # Identify numeric columns
    numeric_cols = billionaires_df.select_dtypes(include=[np.number]).columns.tolist()
    
    # Calculate correlation matrix for numeric columns
    correlation_matrix = billionaires_df[numeric_cols].corr()
    
    # Plot the heatmap
    plt.figure(figsize=(12, 10))
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
    plt.title('Correlation Heatmap')
    plt.show()
    
    
    #Wealth Distribution Analysis
    
    #1. Which sector of the economy produces the most billionaires on average?
    # Step 1: Group by 'category' and calculate the count of billionaires in each sector
    billionaires_per_sector = billionaires_df.groupby('category')['personName'].count()
    
    # Step 2: Calculate the total wealth in each sector
    total_wealth_per_sector = billionaires_df.groupby('category')['finalWorth'].sum()
    
    # Step 3: Calculate the average wealth per billionaire in each sector
    average_wealth_per_billionaire = total_wealth_per_sector / billionaires_per_sector
    
    # Step 4: Combine the results into a DataFrame for better presentation
    sector_analysis_df = pd.DataFrame({
    'Billionaires Count': billionaires_per_sector,
    'Total Wealth': total_wealth_per_sector,
    'Average Wealth per Billionaire': average_wealth_per_billionaire
    })
    
    # Sort the DataFrame by 'Billionaires Count' in descending order
    sector_analysis_df = sector_analysis_df.sort_values(by='Billionaires Count', ascending=False)
    
    # Step 5: Print the detailed analysis
    print("Detailed Analysis of Billionaires by Sector:")
    print(sector_analysis_df)
    
    # Visualization: Bar plot for Average Wealth per Billionaire
    plt.figure(figsize=(12, 8))
    sns.barplot(x='Average Wealth per Billionaire', y=sector_analysis_df.index, data=sector_analysis_df, palette='viridis')
    plt.title('Average Wealth per Billionaire by Sector')
    plt.xlabel('Average Wealth per Billionaire')
    plt.ylabel('Sector')
    plt.show()
    
    
    #2. How much money do billionaires have overall in each nation?   
    # Step 1: Group by 'country' and calculate the total wealth in each nation
    total_wealth_per_nation = billionaires_df.groupby('country')['finalWorth'].sum()
    
    # Step 2: Combine the results into a DataFrame for better presentation
    nation_wealth_df = pd.DataFrame({
    'Total Wealth': total_wealth_per_nation
    }).sort_values(by='Total Wealth', ascending=False)
    
    # Step 3: Print the results
    print("Total Wealth of Billionaires by Nation:")
    print(nation_wealth_df)
    
    import plotly.express as px
    
    # Assuming 'nation_wealth_df' is your DataFrame with total wealth per nation
    
    # Create a treemap
    fig = px.treemap(nation_wealth_df.reset_index(), path=['country'], values='Total Wealth',
         title='Total Wealth Distribution of Billionaires in Each Nation',
         color='Total Wealth', color_continuous_scale='viridis')
    
    # Show the treemap
    fig.show()
    
    #3. What is the largest nation in terms of billionaires, and what sectors do they primarily work in?
        
    # Step 1: Identify the largest nation in terms of billionaires
    largest_nation = billionaires_df.groupby('country')['personName'].count().idxmax()
    
    # Step 2: Filter the data for the largest nation
    largest_nation_data = billionaires_df[billionaires_df['country'] == largest_nation]
    
    # Step 3: Identify the sectors that billionaires in the largest nation primarily work in
    largest_nation_sectors = largest_nation_data.groupby('category')['personName'].count().sort_values(ascending=False)
    
    # Step 4: Combine the results into a DataFrame for better presentation
    largest_nation_analysis_df = pd.DataFrame({
    'Billionaires Count': largest_nation_sectors
    }).sort_values(by='Billionaires Count', ascending=False)
    
    # Step 5: Print the results
    print(f"The largest nation in terms of billionaires is {largest_nation}")
    print("\nSectors primarily worked in by billionaires in the largest nation:")
    print(largest_nation_analysis_df)
    
    # Visualization: Point plot for Sectors Primarily Worked in by Billionaires in the Largest Nation
    plt.figure(figsize=(12, 8))
    sns.pointplot(x='Billionaires Count', y=largest_nation_analysis_df.index, data=largest_nation_analysis_df, join=False, palette='viridis')
    plt.title(f'Sectors Primarily Worked in by Billionaires in {largest_nation}')
    plt.xlabel('Billionaires Count')
    plt.ylabel('Sector')
    plt.show()
    
    #4. How many billionaires are men and how many are women, and how has that ratio changed over time?
    
    # Convert 'birthDate' column to datetime format
    billionaires_df['birthDate'] = pd.to_datetime(billionaires_df['birthDate'], errors='coerce')
    
    # Extract 'gender', 'birthYear', and 'rank' columns
    gender_birth_rank = billionaires_df[['gender', 'birthYear', 'rank']]
    
    # Filter out rows with missing 'gender' or 'birthYear' values
    gender_birth_rank = gender_birth_rank.dropna(subset=['gender', 'birthYear'])
    
    # Group by 'birthYear' and 'gender' to count the number of billionaires
    gender_count = gender_birth_rank.groupby(['birthYear', 'gender']).size().reset_index(name='count')
    
    # Pivot the table for better visualization
    gender_count_pivot = gender_count.pivot(index='birthYear', columns='gender', values='count').fillna(0)
    
    # Calculate the ratio of male to female billionaires
    gender_count_pivot['Male_to_Female_Ratio'] = gender_count_pivot['M'] / gender_count_pivot['F']
    
    # Calculate the total count of male and female billionaires
    total_male_billionaires = gender_birth_rank[gender_birth_rank['gender'] == 'M']['rank'].count()
    total_female_billionaires = gender_birth_rank[gender_birth_rank['gender'] == 'F']['rank'].count()
    
    print(f'Total Male Billionaires: {total_male_billionaires}')
    print(f'Total Female Billionaires: {total_female_billionaires}')
        
    # Plot the graph
    plt.figure(figsize=(12, 8))
    sns.lineplot(data=gender_count_pivot, x='birthYear', y='Male_to_Female_Ratio', color='blue')
    plt.title('Male to Female Billionaires Ratio Over Time')
    plt.xlabel('Birth Year')
    plt.ylabel('Male to Female Ratio')
    plt.show()
    
    #5. Which city or nation has created the most billionaires?
    
    str_return = visualize_top_billionaires_locations(billionaires_df)
    print(str_return)

    #6. How many billionaires are self-made?
    
    visualize_self_made_billionaires(billionaires_df)     
    
    #7. Do self-made billionaires and those who inherited riches differ in terms of average wealth?
    
    visualize_average_wealth(billionaires_df)

    
if __name__=="__main__": 
    main() 
```
